```python
from math import *
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
```

##  Taylorapproximation von $\sin(x)$

```python
def sin_deriv(x,n):
    if n%2==0:
        return (-1)**(n/2)*np.sin(x)
    elif n%2==1:
        return (-1)**((n-1)/2)*np.cos(x)
def Tn(x,n,a=0):
    #n-th Taylor polynomial at a
    return np.sum([sin_deriv(a,i)*(x-a)**i/factorial(i) for i in range(n+1)])
Tn = np.vectorize(Tn)
```

```python
fig = go.Figure()

#Sinus data
t = np.arange(-np.pi, np.pi, 0.01)
sin_graph = np.sin(t)

fig.add_trace(go.Scatter(
    x=t,
    y=sin_graph,
    mode='lines',
    name='sin(x)'
))
fig.add_trace(go.Scatter(x=[0],y=[0],mode='markers',showlegend=False,
                         marker=dict(size=10)))

for deg in range(3):
    fig.add_trace(go.Scatter(
                    visible='legendonly', mode='lines', name='$T_{}$'.format(2*deg+1), x=t, y=Tn(t,2*deg+1),
    ))
fig.update_yaxes(range=[-2, 2])
fig.show()
```
