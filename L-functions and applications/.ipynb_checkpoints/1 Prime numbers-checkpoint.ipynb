## Exercise 1) 
Write a function that returns a list of all prime numbers below x.

<font color ="green"> Allowed functions: Only basic Python functions and sqrt </font>

```python
#Fill in the code below
def prime_list(n):
    if n<=2:
        return [] #That's the empty list
    if n==2:
        return [2]
    L = []
    for i in range(2,n):
        L.append(i)
    return L
```

### <font color ="teal"> Compare your function to the sage inbuilt function primes(n) </font>
From now on we will use the sage inbuilt function

```python
timeit('primes(100000)', preparse=None, number=1, repeat=1, precision=3)
```

```python
%timeit prime_list(1000)
```
